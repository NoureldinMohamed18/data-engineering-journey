```python
from flask import Flask, request, jsonify
# ↑ Flask = المكتبة الأساسية لبناء الـ API
# ↑ request = بيحتوي على بيانات الطلب (JSON, headers, etc.)
# ↑ jsonify = بيحوّل Python dict لـ JSON response

app = Flask(__name__)                # ← إنشاء التطبيق
# ↑ __name__ = اسم الملف الحالي (يخلّي Flask يعرف مكان الـ app)

# قاعدة بيانات وهمية في الذاكرة
users = [{"id": 1, "name": "Ahmed", "email": "ahmed@test.com"}]
# ↑ List of dictionaries = بيانات مؤقتة لحد ما نربط بـ DB حقيقية

# ═══════════════════════════════════════
# GET /users  →  جيب كل الـ users
# ═══════════════════════════════════════
@app.route('/users', methods=['GET'])
# ↑ @app.route = Decorator بيحدد الـ URL والـ Method
# ↑ '/users' = الـ Endpoint
# ↑ methods=['GET'] = الطريقة المسموحة
def get_users():
    return jsonify(users)            # ← بيرجّع الـ List كـ JSON

# ═══════════════════════════════════════
# GET /users/<id>  →  جيب user واحد
# ═══════════════════════════════════════
@app.route('/users/<int:user_id>', methods=['GET'])
# ↑ <int:user_id> = Parameter من الـ URL، لازم يكون رقم
def get_user(user_id):
    # next() = بيجيب أول عنصر مطابق، لو مفيش بيرجّع None
    user = next((u for u in users if u['id'] == user_id), None)
    if user:
        return jsonify(user)         # ← 200 OK (افتراضي)
    return jsonify({"error": "Not found"}), 404
    # ↑ 404 = Status code "Not Found"

# ═══════════════════════════════════════
# POST /users  →  أضف user جديد
# ═══════════════════════════════════════
@app.route('/users', methods=['POST'])
def create_user():
    data = request.get_json()        # ← بياخد الـ JSON من body الطلب
    new_user = {
        "id": len(users) + 1,        # ← ID تلقائي = عدد الـ users + 1
        "name": data['name'],        # ← بياخد الاسم من الـ JSON
        "email": data['email']       # ← بياخد الإيميل من الـ JSON
    }
    users.append(new_user)           # ← بيضيف للـ List
    return jsonify(new_user), 201    # ← 201 = Created

# ═══════════════════════════════════════
# PUT /users/<id>  →  حدّث user كامل
# ═══════════════════════════════════════
@app.route('/users/<int:user_id>', methods=['PUT'])
def update_user(user_id):
    data = request.get_json()
    user = next((u for u in users if u['id'] == user_id), None)
    if user:
        user['name'] = data['name']  # ← بيغيّر الاسم
        user['email'] = data['email']# ← بيغيّر الإيميل
        return jsonify(user)
    return jsonify({"error": "Not found"}), 404

# ═══════════════════════════════════════
# DELETE /users/<id>  →  امسح user
# ═══════════════════════════════════════
@app.route('/users/<int:user_id>', methods=['DELETE'])
def delete_user(user_id):
    global users                     # ← عشان نعدّل الـ List من جوه Function
    users = [u for u in users if u['id'] != user_id]
    # ↑ List comprehension = بيحتفظ بالـ users اللي ID مش زي المطلوب
    return jsonify({"message": "Deleted"})

# ═══════════════════════════════════════
# تشغيل التطبيق
# ═══════════════════════════════════════
if __name__ == '__main__':
    app.run(debug=True)
    # ↑ debug=True = لو فيه خطأ بيظهر تفاصيله (بس للتطوير بس!)
    # ↑ يشتغل على http://localhost:5000
```

```python
import requests                      # ← عشان نجيب الـ HTML
from bs4 import BeautifulSoup        # ← عشان نفكّك الـ HTML

# 1. جيب صفحة الويب
url = "https://quotes.toscrape.com/"
response = requests.get(url)         # ← بيجيب الـ HTML كـ text

# 2. فكّك الـ HTML
soup = BeautifulSoup(response.content, 'html.parser')
# ↑ response.content = الـ HTML الخام (bytes)
# ↑ 'html.parser' = المحلل اللي هيفكّك الـ HTML

# 3. دور على كل الـ quotes
quotes = soup.find_all('div', class_='quote')
# ↑ find_all() = بيجيب كل العناصر المطابقة
# ↑ 'div' = نوع الـ HTML tag
# ↑ class_='quote' = الـ CSS class (class_ مش class عشان class كلمة محجوزة في Python)

for quote in quotes:
    # 4. دور على الـ text جوه كل quote
    text = quote.find('span', class_='text').text
    # ↑ .find() = بيجيب أول عنصر مطابق
    # ↑ .text = بيجيب النص جوه الـ tag (بدون الـ HTML tags)
    
    # 5. دور على الـ author
    author = quote.find('small', class_='author').text
    
    print(f'"{text}" - {author}')
```

```python
from airflow import DAG                                    # ← الـ DAG نفسه
from airflow.operators.python import PythonOperator        # ← Task بيشتغل Python function
from airflow.operators.empty import EmptyOperator          # ← Task فاضي (للـ branching)
from datetime import datetime, timedelta                   # ← التواريخ والتوقيت

# ═══════════════════════════════════════
# Default Arguments (إعدادات افتراضية لكل Tasks)
# ═══════════════════════════════════════
default_args = {
    'owner': 'airflow',           # ← مالك الـ DAG
    'depends_on_past': False,     # ← Task مش معتمد على اللي قبله
    'email_on_failure': False,    # ← متبعتش إيميل لو فشل
    'email_on_retry': False,      # ← متبعتش إيميل لو حاول تاني
    'retries': 1,                 # ← عدد المحاولات لو فشل
    'retry_delay': timedelta(minutes=5),  # ← الانتظار بين المحاولات
}

# ═══════════════════════════════════════
# إنشاء الـ DAG
# ═══════════════════════════════════════
with DAG(
    'weather_dashboard_pipeline',    # ← ID فريد للـ DAG
    default_args=default_args,
    description='Weather data pipeline',
    schedule_interval=timedelta(hours=1),  # ← يشتغل كل ساعة
    start_date=datetime(2024, 1, 1),       # ← تاريخ البداية
    catchup=False,                         # ← متشغلش الماضي اللي فات
) as dag:

    # ═══════════════════════════════════
    # Task 1: Fetch Weather
    # ═══════════════════════════════════
    def fetch_weather():
        import requests
        response = requests.get('https://api.weather.com/forecast')
        data = response.json()
        with open('/tmp/weather.json', 'w') as f:
            import json
            json.dump(data, f)           # ← يحفظ الـ JSON في ملف
        print("Weather data fetched!")

    fetch_task = PythonOperator(
        task_id='fetch_weather_forecast',  # ← ID فريد للـ Task
        python_callable=fetch_weather,     # ← الـ Function اللي هتتنفّذ
    )

    # ═══════════════════════════════════
    # Task 2: Clean Data
    # ═══════════════════════════════════
    def clean_data():
        import json
        with open('/tmp/weather.json', 'r') as f:
            data = json.load(f)
        cleaned = {"temperature": data['temp'], "humidity": data['humidity']}
        with open('/tmp/cleaned_weather.json', 'w') as f:
            json.dump(cleaned, f)
        print("Data cleaned!")

    clean_task = PythonOperator(
        task_id='clean_forecast_data',
        python_callable=clean_data,
    )

    # ═══════════════════════════════════
    # Task 3: Push to Dashboard
    # ═══════════════════════════════════
    def push_to_dashboard():
        print("Data pushed to dashboard!")

    push_task = PythonOperator(
        task_id='push_data_to_dashboard',
        python_callable=push_to_dashboard,
    )

    # ═══════════════════════════════════
    # الـ Dependencies (الترتيب)
    # ═══════════════════════════════════
    fetch_task >> clean_task >> push_task
    # ↑ >> = يعني "بعد ما يخلص" (depends on)
    # ↑ fetch → clean → push (خطي)
```

```python
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime, timedelta
import requests
import sqlite3

# ═══════════════════════════════════════
# DAG يجيب بيانات من API ويحفظها كل يوم
# ═══════════════════════════════════════

default_args = {
    'owner': 'data_engineer',
    'depends_on_past': False,
    'email_on_failure': False,
    'retries': 1,
    'retry_delay': timedelta(minutes=5),
}

with DAG(
    'github_users_pipeline',
    default_args=default_args,
    description='جلب بيانات GitHub users يومياً',
    schedule=timedelta(days=1),  # يشتغل كل يوم
    start_date=datetime(2024, 1, 1),
    catchup=False,
) as dag:

    # Task 1: جلب البيانات من API
    def fetch_github_users():
        response = requests.get(
            "https://api.github.com/users",
            params={"per_page": 10},
            timeout=10
        )
        users = response.json()
        print(f"✅ تم جلب {len(users)} users")
        return users

    fetch_task = PythonOperator(
        task_id='fetch_users',
        python_callable=fetch_github_users,
    )

    # Task 2: تنظيف البيانات
    def clean_users(**context):
        users = context['ti'].xcom_pull(task_ids='fetch_users')
        
        cleaned = []
        for user in users:
            cleaned.append({
                "id": user["id"],
                "name": user["login"],
                "email": user.get("email") or f"{user['login']}@github.com",
                "company": user.get("company") or "Independent"
            })
        
        print(f"✅ تم تنظيف {len(cleaned)} users")
        return cleaned

    clean_task = PythonOperator(
        task_id='clean_users',
        python_callable=clean_users,
    )

    # Task 3: حفظ في Database
    def save_to_db(**context):
        users = context['ti'].xcom_pull(task_ids='clean_users')
        
        conn = sqlite3.connect("github_users.db")
        cursor = conn.cursor()
        
        for user in users:
            cursor.execute("""
                INSERT OR REPLACE INTO users (id, name, email, company)
                VALUES (?, ?, ?, ?)
            """, (user["id"], user["name"], user["email"], user["company"]))
        
        conn.commit()
        conn.close()
        print(f"✅ تم حفظ {len(users)} users في Database")

    save_task = PythonOperator(
        task_id='save_to_db',
        python_callable=save_to_db,
    )

    # الترتيب: جلب → تنظيف → حفظ
    fetch_task >> clean_task >> save_task
```

```python

# ═══════════════════════════════════════════════════════════
# GIT & GITHUB - شرح عملي بالكود
# ═══════════════════════════════════════════════════════════

print("=" * 70)
print("🐙 GIT & GITHUB - من الصفر للاحتراف")
print("=" * 70)

# ─── محاكاة لـ Git Workflow ───
print("""
📋 GIT WORKFLOW (الخطوات اللي هتمشي عليها دايماً):

┌─────────────┐    ┌─────────────┐    ┌─────────────┐    ┌─────────────┐
│  Working    │───►│   Staging   │───►│  Commit     │───►│   Push      │
│  Directory  │    │   (Index)   │    │  (Local)    │    │  (Remote)   │
└─────────────┘    └─────────────┘    └─────────────┘    └─────────────┘
     │                   │                  │                  │
     │                   │                  │                  │
  git add .          git commit        git push          GitHub
                     -m "msg"          origin main
""")

# ─── محاكاة Git Commands بـ Python ───
print("\n💻 محاكاة أوامر Git بـ Python:")
print("-" * 70)

class GitSimulator:
    """محاكاة لـ Git"""
    
    def __init__(self, repo_name):
        self.repo_name = repo_name
        self.working_dir = {}  # الملفات اللي شغال عليها
        self.staging = {}      # الملفات اللي هتتعملها commit
        self.commits = []      # كل الـ commits
        self.branches = {"main": []}
        self.current_branch = "main"
        self.remote = None
        
        print(f"🆕 Initialized empty Git repository in ./{repo_name}/")
    
    def create_file(self, filename, content):
        """إنشاء ملف جديد"""
        self.working_dir[filename] = content
        print(f"📄 Created: {filename}")
    
    def status(self):
        """git status"""
        print(f"\n📊 git status")
        print(f"   On branch {self.current_branch}")
        
        untracked = [f for f in self.working_dir if f not in self.staging]
        staged = [f for f in self.staging if f not in [c.get('files', []) for c in self.commits]]
        
        if untracked:
            print(f"   Untracked files: {untracked}")
        if staged:
            print(f"   Changes to be committed: {staged}")
        if not untracked and not staged:
            print(f"   nothing to commit, working tree clean")
    
    def add(self, filename="."):
        """git add"""
        if filename == ".":
            self.staging.update(self.working_dir)
            print(f"✅ git add . → Added {len(self.working_dir)} file(s) to staging")
        else:
            self.staging[filename] = self.working_dir[filename]
            print(f"✅ git add {filename} → Added to staging")
    
    def commit(self, message):
        """git commit"""
        if not self.staging:
            print("❌ Nothing to commit")
            return
        
        commit = {
            "id": f"{len(self.commits)+1:04x}",
            "message": message,
            "files": dict(self.staging),
            "branch": self.current_branch,
            "timestamp": "2024-07-03 10:00:00"
        }
        self.commits.append(commit)
        self.branches[self.current_branch].append(commit)
        self.staging = {}
        print(f"✅ git commit -m \"{message}\"")
        print(f"   [{commit['id']}] {message}")
    
    def log(self):
        """git log"""
        print(f"\n📜 git log")
        for c in reversed(self.commits):
            print(f"   commit {c['id']}")
            print(f"   Author: Data Engineer <de@example.com>")
            print(f"   Date: {c['timestamp']}")
            print(f"   {c['message']}")
            print()
    
    def branch(self, name=None):
        """git branch"""
        if name:
            self.branches[name] = list(self.branches[self.current_branch])
            print(f"✅ git branch {name}")
        else:
            print(f"\n🌿 Branches:")
            for b in self.branches:
                marker = "*" if b == self.current_branch else " "
                print(f"   {marker} {b}")
    
    def checkout(self, name):
        """git checkout"""
        if name not in self.branches:
            # Create and checkout
            self.branches[name] = list(self.branches[self.current_branch])
            print(f"✅ git checkout -b {name}")
        else:
            print(f"✅ git checkout {name}")
        self.current_branch = name
    
    def merge(self, branch_name):
        """git merge"""
        print(f"✅ git merge {branch_name}")
        print(f"   Merged {branch_name} into {self.current_branch}")
    
    def remote_add(self, url):
        """git remote add origin"""
        self.remote = url
        print(f"✅ git remote add origin {url}")
    
    def push(self):
        """git push"""
        if not self.remote:
            print("❌ No remote configured")
            return
        print(f"✅ git push -u origin {self.current_branch}")
        print(f"   Pushed {len(self.commits)} commits to {self.remote}")

# ═══════════════════════════════════════════════════════════
# 🎯 التطبيق العملي: بناء أول مشروع Data Engineering
# ═══════════════════════════════════════════════════════════

print("\n" + "=" * 70)
print("🎯 PROJECT: بناء أول Repo على GitHub")
print("=" * 70)

# 1. Initialize Repo
repo = GitSimulator("data-engineering-projects")

# 2. Create files
print("\n📁 Creating project files...")
repo.create_file("README.md", """# Data Engineering Projects

## About
This repository contains my data engineering projects.

## Projects
1. Employee Management System
2. E-commerce Data Warehouse

## Skills
- Python
- SQL
- Data Modeling
- ETL Pipelines
""")

repo.create_file("employee_management.py", "# Employee Management System\n# TODO: Implement ETL pipeline")
repo.create_file("requirements.txt", "pandas\nnumpy\nmatplotlib")

# 3. Check status
repo.status()

# 4. Add files
repo.add(".")

# 5. Commit
repo.commit("feat: initial project setup with README and structure")

# 6. Create feature branch
repo.checkout("feature/etl-pipeline")

# 7. Add new feature
repo.create_file("etl_pipeline.py", "# ETL Pipeline Implementation\n# Extract -> Transform -> Load")
repo.add("etl_pipeline.py")
repo.commit("feat: add basic ETL pipeline structure")

# 8. Switch back to main and merge
repo.checkout("main")
repo.merge("feature/etl-pipeline")

# 9. Connect to GitHub
repo.remote_add("https://github.com/yourusername/data-engineering-projects.git")

# 10. Push
repo.push()

# 11. View log
repo.log()

print("\n" + "=" * 70)
print("✅ Git Workflow Complete!")
print("=" * 70)
```

```python
import csv

def clean_name(name):
    return name.strip().title()

def clean_department(dept):
    return dept.strip().lower()

def parse_status(status):
    return status.strip().lower() == "yes"

def calculate_tax(salary, rate=0.14):
    return round(salary * rate, 2)

def process_employees(input_file, output_file):
    with open(input_file, "r") as f:
        reader = csv.DictReader(f)
        rows = list(reader)
    
    clean_rows = []
    for row in rows:
        salary = float(row["salary"])
        tax = calculate_tax(salary)
        
        clean_rows.append({
            "id": int(row["id"]),
            "name": clean_name(row["name"]),
            "email": row["email"].strip().lower(),
            "salary": salary,
            "tax": tax,
            "net_salary": round(salary - tax, 2),
            "department": clean_department(row["department"]),
            "is_active": parse_status(row["status"])
        })
    
    with open(output_file, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=clean_rows[0].keys())
        writer.writeheader()
        writer.writerows(clean_rows)
    
    print(f"✅ Processed {len(clean_rows)} employees")
    print(f"💾 Saved to: {output_file}")

# Run
process_employees("raw_employees.csv", "clean_employees.csv")
```

```python

import sqlite3

# ═══════════════════════════════════════════════════════════
# SQL REVIEW FOR DATA ENGINEERING
# ═══════════════════════════════════════════════════════════

print("=" * 70)
print("🗄️ SQL REVIEW FOR DATA ENGINEERING")
print("=" * 70)

# ─── 1. إنشاء Database و Tables ───
print("\n🏗️ 1. CREATE DATABASE & TABLES")
print("-" * 70)

conn = sqlite3.connect(":memory:")  # Database في الذاكرة (للتجربة)
cursor = conn.cursor()

# Create tables
cursor.executescript("""
-- Dimension Table: customers
CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    city TEXT,
    segment TEXT
);

-- Dimension Table: products
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    category TEXT,
    price REAL NOT NULL,
    cost REAL
);

-- Fact Table: orders
CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    order_date TEXT,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id)
);
""")

print("✅ Created: customers, products, orders")

# ─── 2. INSERT DATA ───
print("\n📥 2. INSERT DATA")
print("-" * 70)

cursor.executemany("INSERT INTO customers VALUES (?, ?, ?, ?, ?)", [
    (1, "Ahmed", "ahmed@mail.com", "Cairo", "Premium"),
    (2, "Sara", "sara@mail.com", "Alex", "Standard"),
    (3, "Omar", "omar@mail.com", "Giza", "Standard"),
    (4, "Fatima", "fatima@mail.com", "Cairo", "Premium"),
])

cursor.executemany("INSERT INTO products VALUES (?, ?, ?, ?, ?)", [
    (101, "Laptop", "Electronics", 1200, 900),
    (102, "Mouse", "Electronics", 25, 15),
    (103, "Chair", "Furniture", 150, 100),
    (104, "Mug", "Kitchen", 12, 5),
])

cursor.executemany("INSERT INTO orders VALUES (?, ?, ?, ?, ?)", [
    (1, 1, 101, 2, "2024-01-15"),
    (2, 1, 102, 3, "2024-01-15"),
    (3, 2, 101, 1, "2024-01-16"),
    (4, 3, 103, 2, "2024-01-17"),
    (5, 4, 101, 1, "2024-01-18"),
    (6, 2, 104, 5, "2024-01-18"),
])

conn.commit()
print("✅ Inserted: 4 customers, 4 products, 6 orders")

# ─── 3. SELECT - Basic ───
print("\n🔍 3. SELECT - Basic Queries")
print("-" * 70)

print("\n📌 SELECT * FROM customers:")
cursor.execute("SELECT * FROM customers")
for row in cursor.fetchall():
    print(f"  {row}")

print("\n📌 SELECT name, email FROM customers:")
cursor.execute("SELECT name, email FROM customers")
for row in cursor.fetchall():
    print(f"  {row}")

print("\n📌 SELECT * FROM customers WHERE segment = 'Premium':")
cursor.execute("SELECT * FROM customers WHERE segment = 'Premium'")
for row in cursor.fetchall():
    print(f"  {row}")

# ─── 4. JOIN - الربط (أهم حاجة في DE!) ───
print("\n\n🔗 4. JOIN - ربط الجداول (Star Schema)")
print("-" * 70)

print("\n📌 INNER JOIN: orders + customers + products")
cursor.execute("""
    SELECT 
        o.order_id,
        c.name AS customer,
        p.name AS product,
        o.quantity,
        p.price,
        o.quantity * p.price AS total
    FROM orders o
    INNER JOIN customers c ON o.customer_id = c.customer_id
    INNER JOIN products p ON o.product_id = p.product_id
""")
print(f"{'Order':<8} {'Customer':<10} {'Product':<10} {'Qty':<5} {'Price':<8} {'Total':<8}")
print("-" * 55)
for row in cursor.fetchall():
    print(f"{row[0]:<8} {row[1]:<10} {row[2]:<10} {row[3]:<5} ${row[4]:<7.0f} ${row[5]:<7.2f}")

# ─── 5. GROUP BY - التجميع ───
print("\n\n📊 5. GROUP BY - التجميع (Aggregations)")
print("-" * 70)

print("\n📌 Sales by Customer:")
cursor.execute("""
    SELECT 
        c.name,
        COUNT(o.order_id) AS order_count,
        SUM(o.quantity * p.price) AS total_spent
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    JOIN products p ON o.product_id = p.product_id
    GROUP BY c.customer_id
    ORDER BY total_spent DESC
""")
print(f"{'Customer':<10} {'Orders':<8} {'Total Spent':<12}")
print("-" * 35)
for row in cursor.fetchall():
    print(f"{row[0]:<10} {row[1]:<8} ${row[2]:<11.2f}")

print("\n📌 Sales by Category:")
cursor.execute("""
    SELECT 
        p.category,
        COUNT(DISTINCT o.order_id) AS orders,
        SUM(o.quantity) AS total_qty,
        SUM(o.quantity * p.price) AS revenue
    FROM products p
    JOIN orders o ON p.product_id = o.product_id
    GROUP BY p.category
    ORDER BY revenue DESC
""")
print(f"{'Category':<15} {'Orders':<8} {'Qty':<8} {'Revenue':<12}")
print("-" * 50)
for row in cursor.fetchall():
    print(f"{row[0]:<15} {row[1]:<8} {row[2]:<8} ${row[3]:<11.2f}")

# ─── 6. Window Functions (Advanced) ───
print("\n\n🪟 6. WINDOW FUNCTIONS (Advanced DE)")
print("-" * 70)

print("\n📌 RANK: Top customers by spending")
cursor.execute("""
    SELECT 
        c.name,
        SUM(o.quantity * p.price) AS total,
        RANK() OVER (ORDER BY SUM(o.quantity * p.price) DESC) AS rank
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    JOIN products p ON o.product_id = p.product_id
    GROUP BY c.customer_id
""")
print(f"{'Rank':<6} {'Customer':<10} {'Total':<12}")
print("-" * 30)
for row in cursor.fetchall():
    print(f"{row[2]:<6} {row[0]:<10} ${row[1]:<11.2f}")

print("\n📌 Running Total: Cumulative sales by date")
cursor.execute("""
    SELECT 
        o.order_date,
        o.quantity * p.price AS daily_sale,
        SUM(o.quantity * p.price) OVER (ORDER BY o.order_date) AS running_total
    FROM orders o
    JOIN products p ON o.product_id = p.product_id
    ORDER BY o.order_date
""")
print(f"{'Date':<12} {'Daily':<10} {'Running Total':<15}")
print("-" * 40)
for row in cursor.fetchall():
    print(f"{row[0]:<12} ${row[1]:<9.2f} ${row[2]:<14.2f}")

# ─── 7. CTE - Common Table Expressions ───
print("\n\n📋 7. CTE - تنظيم الـ Queries المعقدة")
print("-" * 70)

print("\n📌 CTE: Customer spending with averages")
cursor.execute("""
    WITH customer_totals AS (
        SELECT 
            c.customer_id,
            c.name,
            SUM(o.quantity * p.price) AS total_spent
        FROM customers c
        JOIN orders o ON c.customer_id = o.customer_id
        JOIN products p ON o.product_id = p.product_id
        GROUP BY c.customer_id
    )
    SELECT 
        name,
        total_spent,
        AVG(total_spent) OVER () AS avg_spending,
        total_spent - AVG(total_spent) OVER () AS diff_from_avg
    FROM customer_totals
    ORDER BY total_spent DESC
""")
print(f"{'Customer':<10} {'Spent':<10} {'Avg':<10} {'Diff':<10}")
print("-" * 45)
for row in cursor.fetchall():
    print(f"{row[0]:<10} ${row[1]:<9.2f} ${row[2]:<9.2f} ${row[3]:<9.2f}")

# ─── 8. Indexing for Performance ───
print("\n\n📇 8. INDEXING - تحسين الأداء")
print("-" * 70)

# Before index
cursor.execute("EXPLAIN QUERY PLAN SELECT * FROM orders WHERE customer_id = 1")
print("📌 Query Plan BEFORE index:")
for row in cursor.fetchall():
    print(f"  {row}")

# Create index
cursor.execute("CREATE INDEX idx_orders_customer ON orders(customer_id)")
print("\n✅ Created index: idx_orders_customer")

# After index
cursor.execute("EXPLAIN QUERY PLAN SELECT * FROM orders WHERE customer_id = 1")
print("\n📌 Query Plan AFTER index:")
for row in cursor.fetchall():
    print(f"  {row}")

conn.close()

print("\n" + "=" * 70)
print("✅ SQL Review Complete!")
print("=" * 70)
```

```python
import sqlite3

def create_database():
    conn = sqlite3.connect("sales.db")
    cursor = conn.cursor()
    
    cursor.executescript("""
        CREATE TABLE IF NOT EXISTS products (
            product_id INTEGER PRIMARY KEY,
            name TEXT,
            category TEXT,
            price REAL
        );
        
        CREATE TABLE IF NOT EXISTS sales (
            sale_id INTEGER PRIMARY KEY,
            product_id INTEGER,
            quantity INTEGER,
            sale_date TEXT,
            FOREIGN KEY (product_id) REFERENCES products(product_id)
        );
    """)
    
    cursor.executemany("INSERT OR IGNORE INTO products VALUES (?, ?, ?, ?)", [
        (1, "Laptop", "Electronics", 1200),
        (2, "Mouse", "Electronics", 25),
        (3, "Chair", "Furniture", 150),
    ])
    
    cursor.executemany("INSERT INTO sales VALUES (?, ?, ?, ?)", [
        (1, 1, 2, "2024-01-01"),
        (2, 2, 5, "2024-01-01"),
        (3, 1, 1, "2024-01-02"),
        (4, 3, 3, "2024-01-02"),
    ])
    
    conn.commit()
    return conn

def get_sales_by_category(conn):
    cursor = conn.cursor()
    cursor.execute("""
        SELECT p.category, SUM(s.quantity * p.price) as revenue
        FROM products p
        JOIN sales s ON p.product_id = s.product_id
        GROUP BY p.category
        ORDER BY revenue DESC
    """)
    return cursor.fetchall()

def get_daily_sales(conn):
    cursor = conn.cursor()
    cursor.execute("""
        SELECT sale_date, SUM(quantity * price) as daily_total
        FROM sales s
        JOIN products p ON s.product_id = p.product_id
        GROUP BY sale_date
        ORDER BY sale_date
    """)
    return cursor.fetchall()

def main():
    conn = create_database()
    
    print("📊 SALES ANALYTICS DASHBOARD\n")
    
    print("By Category:")
    for cat, rev in get_sales_by_category(conn):
        print(f"  {cat}: ${rev:.2f}")
    
    print("\nDaily Sales:")
    for date, total in get_daily_sales(conn):
        print(f"  {date}: ${total:.2f}")
    
    conn.close()

if __name__ == "__main__":
    main()
```

```python

# ═══════════════════════════════════════════════════════════
# NOSQL REVIEW - MongoDB with Python
# ═══════════════════════════════════════════════════════════

print("=" * 70)
print("📦 NOSQL REVIEW - MongoDB with Python")
print("=" * 70)

# ─── محاكاة MongoDB بـ Dictionary ───
print("\n🗂️ MongoDB Document Structure (JSON-like)")
print("-" * 70)

# MongoDB Document = Dict في Python
customer_doc = {
    "_id": 1001,
    "name": "Ahmed Mohamed",
    "email": "ahmed@mail.com",
    "addresses": [  # Nested array (مش موجود في SQL!)
        {"type": "home", "city": "Cairo", "street": "Tahrir"},
        {"type": "work", "city": "Alex", "street": "Corniche"}
    ],
    "orders": [  # Nested documents
        {"order_id": "ORD001", "total": 1200, "items": ["Laptop"]},
        {"order_id": "ORD002", "total": 150, "items": ["Mouse", "Keyboard"]}
    ],
    "metadata": {
        "created_at": "2024-01-01",
        "source": "website",
        "tags": ["premium", "active"]
    }
}

print(f"Customer Document:")
print(f"  Name: {customer_doc['name']}")
print(f"  Home Address: {customer_doc['addresses'][0]['city']}")
print(f"  First Order: {customer_doc['orders'][0]['order_id']}")
print(f"  Tags: {customer_doc['metadata']['tags']}")

# ─── محاكاة MongoDB Collection ───
print("\n\n📚 MongoDB Collection (محاكاة)")
print("-" * 70)

class MongoCollection:
    """محاكاة لـ MongoDB Collection"""
    
    def __init__(self, name):
        self.name = name
        self.documents = []
    
    def insert_one(self, doc):
        self.documents.append(doc)
        print(f"✅ Inserted 1 document into {self.name}")
    
    def insert_many(self, docs):
        self.documents.extend(docs)
        print(f"✅ Inserted {len(docs)} documents into {self.name}")
    
    def find(self, query=None):
        if query is None:
            return self.documents
        return [d for d in self.documents if all(d.get(k) == v for k, v in query.items())]
    
    def find_one(self, query):
        results = self.find(query)
        return results[0] if results else None
    
    def update_one(self, query, update):
        for doc in self.documents:
            if all(doc.get(k) == v for k, v in query.items()):
                doc.update(update.get("$set", {}))
                print(f"✅ Updated document in {self.name}")
                return
    
    def delete_one(self, query):
        self.documents = [d for d in self.documents if not all(d.get(k) == v for k, v in query.items())]
        print(f"✅ Deleted document from {self.name}")
    
    def aggregate(self, pipeline):
        """محاكاة بسيطة لـ Aggregation Pipeline"""
        result = self.documents
        for stage in pipeline:
            if "$match" in stage:
                result = [d for d in result if all(d.get(k) == v for k, v in stage["$match"].items())]
            elif "$group" in stage:
                # محاكاة بسيطة
                pass
            elif "$sort" in stage:
                key = list(stage["$sort"].keys())[0]
                reverse = stage["$sort"][key] == -1
                result = sorted(result, key=lambda x: x.get(key, 0), reverse=reverse)
        return result

# نعمل Collection
customers_collection = MongoCollection("customers")

# Insert documents
customers_collection.insert_many([
    {"_id": 1, "name": "Ahmed", "city": "Cairo", "segment": "Premium", "orders_count": 5},
    {"_id": 2, "name": "Sara", "city": "Alex", "segment": "Standard", "orders_count": 3},
    {"_id": 3, "name": "Omar", "city": "Cairo", "segment": "Premium", "orders_count": 8},
    {"_id": 4, "name": "Fatima", "city": "Giza", "segment": "Basic", "orders_count": 1},
])

# Find all
print(f"\n📌 All Customers:")
for c in customers_collection.find():
    print(f"  {c['name']} ({c['segment']}) - {c['orders_count']} orders")

# Find with query
print(f"\n📌 Premium Customers in Cairo:")
premium_cairo = customers_collection.find({"segment": "Premium", "city": "Cairo"})
for c in premium_cairo:
    print(f"  {c['name']}: {c['orders_count']} orders")

# Update
print(f"\n📌 Update Sara to Premium:")
customers_collection.update_one({"name": "Sara"}, {"$set": {"segment": "Premium"}})
print(f"  Sara's new segment: {customers_collection.find_one({'name': 'Sara'})['segment']}")

# ─── Comparison: SQL vs NoSQL ───
print("\n\n⚖️ SQL vs NoSQL Comparison")
print("-" * 70)

comparison = """
┌─────────────────┬─────────────────────┬─────────────────────┐
│ Feature         │ SQL                 │ NoSQL               │
├─────────────────┼─────────────────────┼─────────────────────┤
│ Structure       │ Tables (Rows/Cols)  │ Documents (JSON)    │
│ Schema          │ Fixed (Rigid)       │ Flexible            │
│ Scaling         │ Vertical (Bigger HW)│ Horizontal (More)   │
│ Joins           │ Yes (Easy)          │ No (Embedded)       │
│ Transactions    │ ACID (Strong)       │ BASE (Eventual)     │
│ Best For        │ Structured Data     │ Unstructured Data   │
│ Examples        │ PostgreSQL, MySQL   │ MongoDB, Cassandra  │
└─────────────────┴─────────────────────┴─────────────────────┘

💡 In Data Engineering:
• SQL = Data Warehouse (Analytics)
• NoSQL = Data Lake (Raw Data)
• Both = Modern Data Architecture
"""
print(comparison)

# ─── When to use what? ───
print("\n🎯 When to use SQL vs NoSQL in DE:")
print("-" * 70)

use_cases = {
    "SQL (PostgreSQL)": [
        "Data Warehouse (Star Schema)",
        "Financial Transactions",
        "Complex Joins & Analytics",
        "ACID Compliance"
    ],
    "NoSQL (MongoDB)": [
        "Raw Data Ingestion (Logs, JSON)",
        "User Profiles (Nested Data)",
        "Real-time Analytics",
        "High Write Throughput"
    ],
    "Both Together": [
        "Lambda Architecture",
        "Data Lake + Data Warehouse",
        "Raw + Processed Data"
    ]
}

for db, cases in use_cases.items():
    print(f"\n📌 {db}:")
    for case in cases:
        print(f"   • {case}")

print("\n" + "=" * 70)
print("✅ NoSQL Review Complete!")
print("=" * 70)
```

```python
import sqlite3
import json

# SQL Database (Data Warehouse)
def create_sql_db():
    conn = sqlite3.connect("warehouse.db")
    cursor = conn.cursor()
    
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS processed_customers (
            customer_id INTEGER PRIMARY KEY,
            name TEXT,
            email TEXT,
            total_orders INTEGER,
            total_spent REAL
        )
    """)
    
    cursor.executemany("INSERT OR IGNORE INTO processed_customers VALUES (?, ?, ?, ?, ?)", [
        (1, "Ahmed", "ahmed@mail.com", 5, 5000.00),
        (2, "Sara", "sara@mail.com", 3, 2500.00),
    ])
    
    conn.commit()
    return conn

# NoSQL (JSON files - simulating MongoDB)
def create_nosql_data():
    raw_data = {
        "customers": [
            {"_id": 1, "name": "Ahmed", "raw_events": ["login", "purchase", "logout"]},
            {"_id": 2, "name": "Sara", "raw_events": ["login", "browse"]},
        ],
        "logs": [
            {"timestamp": "2024-01-01 10:00:00", "event": "page_view", "user_id": 1},
            {"timestamp": "2024-01-01 10:05:00", "event": "purchase", "user_id": 1},
        ]
    }
    
    with open("raw_data.json", "w") as f:
        json.dump(raw_data, f, indent=2)
    
    return raw_data

# ETL: Extract from NoSQL → Transform → Load to SQL
def etl_pipeline():
    print("🔄 ETL Pipeline: NoSQL → SQL\n")
    
    # Extract from NoSQL
    with open("raw_data.json", "r") as f:
        raw_data = json.load(f)
    
    print("📥 Extracted from NoSQL:")
    for customer in raw_data["customers"]:
        print(f"  {customer['name']}: {len(customer['raw_events'])} events")
    
    # Transform
    transformed = []
    for customer in raw_data["customers"]:
        transformed.append({
            "customer_id": customer["_id"],
            "name": customer["name"],
            "event_count": len(customer["raw_events"]),
            "engagement_score": len(customer["raw_events"]) * 10
        })
    
    print("\n🔧 Transformed:")
    for t in transformed:
        print(f"  {t['name']}: Score = {t['engagement_score']}")
    
    # Load to SQL
    conn = sqlite3.connect("warehouse.db")
    cursor = conn.cursor()
    
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS engagement_metrics (
            customer_id INTEGER PRIMARY KEY,
            name TEXT,
            event_count INTEGER,
            engagement_score INTEGER
        )
    """)
    
    for t in transformed:
        cursor.execute("""
            INSERT OR REPLACE INTO engagement_metrics VALUES (?, ?, ?, ?)
        """, (t["customer_id"], t["name"], t["event_count"], t["engagement_score"]))
    
    conn.commit()
    print("\n💾 Loaded to SQL Warehouse")
    
    # Verify
    cursor.execute("SELECT * FROM engagement_metrics")
    print("\n✅ Data in Warehouse:")
    for row in cursor.fetchall():
        print(f"  {row}")
    
    conn.close()

def main():
    create_sql_db()
    create_nosql_data()
    etl_pipeline()

if __name__ == "__main__":
    main()
```
